```python
import numpy as np
import matplotlib.pyplot as mt
t=np.arange(0,10,1)
mt.xlabel('x-axis')
mt.ylabel('y-axis')
mt.title('plot')
mt.plot(t,t,'r--',t,t**2,'g^',t,t**3,'bo')
mt.legend()
mt.show()
```

```python
mt.plot(t,t**5,'g^',label='1st plot')
mt.plot(t,t**3,'r--',label='2nd plot')
mt.grid()
plt.yscale('linear')
plt.xscale('linear')
plt.axis([0,6,0,200])
mt.legend()
mt.show()
```

```python
labels='frog','duck','hen','dog'
sizes=[10,20,30,40]
explode=[0,0,0,0.1]
mt.pie(sizes,labels=labels,explode=explode,shadow=True,autopct='%1.1f%%')
mt.axis('equal')
mt.show()
```
